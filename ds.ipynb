```python
import numpy as np
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
```

# Reading the Data

```python

sentences = pd.read_csv("C:/Users/STSC/Desktop/wana-interview-test-master/wana-interview-test-master/datascience/sentiment_posts.csv")
```

# Testing the Data

```python
sid = SentimentIntensityAnalyzer()
for sentence in sentences["text_original"]:
    print(sentence)
    ss = sid.polarity_scores(sentence)
    ss.pop("compound")
    for k in sorted(ss):
        print('{0}: {1}, '.format(k, ss[k]), end=' ')
    print()
```
